# Statistic with numpy

### Describing a Single Set of Data

# Central Tendencies (중심집중경향) 
### mean
* 평균/대표값/average
* mean != median

### median
* 중앙값/medoid

### quantile
* 백분위

### mode
* 최빈값/frequency

```python
import numpy as np
```

```python
x = np.random.randn(5,7)
print(x)
```

```python
print("mean: ", np.mean(x))
```

```python
print("median: ", np.median(x))
```

```python
print("quantile 0.10(상위 10%) : ", np.percentile(x,0.10))
```

```python
print("quantile 0.10(상위 90%/하위 10%) : ", np.percentile(x,0.90))
```

# dispersion (분산)
### range
* 범위(최대와 최소의 차이)|

### variance
* 분산

### Standard deviation
* 표준편차

### interquartile
* 백분위의 범위

```python
x = np.random.randn(5,7)
print(x)
```

```python
print("range : ", np.ptp(x))
print("range : ", np.max(x)-np.min(x))
```

```python
print("variance : ", np.var(x, ddof=1))
```

```python
print("standard deviation : ", np.std(x, ddof=1))
```

```python
q75, q25 = np.percentile(x,[75,25])
print("interqurtile 25~75(사분변위):", q75-q25)
```
